```python
# Assume that we want to replace the missing values of variable "normalized-losses" by the
#-- the mean value of the variable: the missing value must be replace by the average of the 
#-- entries within that column

import pandas as pd
import numpy as np
df=pd.read_csv("usedcars.csv")
#df.head(20)
#df.dtypes()
#df["normalized-losses"].head()
total=0
count=0
mean=0
for row in df["normalized-losses"]:
    if (row != "?"):
        total=total+int(row)
        count+=1
mean=total/count
print(total)
print(mean)
#df["normalized-losses"].head(20)
df["normalized-losses"] = df["normalized-losses"].replace("?", mean)
df.head()
```

### How to deal with missing data? 
#### 1. Eliminating training examples with missing values.

```python

df.dropna(axis=0)
# axis=0: drop row contains NaN.
```

```python
df.dropna(axis=1)
# axis=1" drop column contains NaN.
```

```python
# only drop rows where all values are NaN
# (return the whole array since we do not have a row with all values NaN)
df.dropna(how='all')
```

```python
# drop rows that have fewer than 4 real values
df.dropna(thresh=4)
```

```python
# only drop rows where NaN appear in specific columns (here: 'C')
df.dropna(subset=['C'])
```

##### Remark: 
Although the removal of missing data seems to be a convenient approach, it also 
comes with certain disadvantages; for example, we may end up removing too many samples, which will make a reliable analysis impossible. Or, if we remove too many feature columns, we will run the risk of losing valuable information that our classifier needs to discriminate between classes. In the next section, we will look at one of the most commonly used alternatives for dealing with missing values: interpolation techniques.

#### 2. Imputing Missing Values:
One of the most common interpolation techniques is **mean imputation**, where we simply relace the missing value with the **mean value of the entire feature column**:

```python
from sklearn.impute import SimpleImputer
import numpy as np
imr = SimpleImputer(missing_values=np.nan, strategy='mean')
# The other options for the strategy parameter are "median" and "most_frequent"
print(df)

imr = imr.fit(df.values)
imputed_data = imr.transform(df.values)
imputed_data
```

Alternatively, we can obtain the same result by using **fillna** method and providing an imputation method as an argument:

```python
df.fillna(df.mean())
```
